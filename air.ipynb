```python
from graphics import Graph3D, Graph2D, ajuste
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

aire = pd.read_csv("Aire1.csv", delimiter=";")
aire.replace([np.inf, -np.inf], np.nan, inplace=True)
aire.dropna(inplace=True)
aire[["theta1", "theta2", "theta3"]] = aire[["theta1", "theta2", "theta3"]] * (np.pi / 180)
aire[["omega1", "omega2", "omega3"]] = aire[["omega1", "omega2", "omega3"]] * (np.pi / 180)

def first_aire(t,gamma,omega,alpha,A):
    return max(aire["theta1"])*np.exp(-gamma * t)*np.cos(omega*t + alpha) + A

def second_aire(t,gamma,omega,alpha,A):
    return max(aire["theta2"])*np.exp(-gamma * t)*np.cos(omega*t + alpha) + A

def third_aire(t,gamma,omega,alpha,A):
    return max(aire["theta3"])*np.exp(-gamma * t)*np.cos(omega*t + alpha) + A

print(max(aire["theta1"]))
print(max(aire["theta2"]))
print(max(aire["theta3"]))
```

```python
param_1=ajuste(
    x_data=aire["t1"], 
    y_data=aire["theta1"], 
    modelo_func=first_aire, 
    x_label="Tiempo (s)", 
    y_label="√Ångulo (Œ∏)", 
    etiqueta=r"$\theta$ vs tiempo", 
    modelo_latex=r"$\theta(t) =  0.24e^{-\gamma t} \cos(\omega t + \alpha) + A$",  # Ejemplo de funci√≥n en LaTeX
    color_mapa="viridis", 
    p0=[0.1, 1, 0, max(aire["theta1"])],  # Valores iniciales para gamma, omega, alpha, y A
    save_path= rf'Graficos\ajuste_grafico1.png'
)
```

```python
param_2=ajuste(
    x_data=aire["t2"], 
    y_data=aire["theta2"], 
    modelo_func=second_aire, 
    x_label="Tiempo (s)", 
    y_label="√Ångulo (Œ∏)", 
    etiqueta=r"$\theta$ vs tiempo", 
    modelo_latex=r"$\theta(t) =  0.21e^{-\gamma t} \cos(\omega t + \alpha) + A$",  # Ejemplo de funci√≥n en LaTeX
    color_mapa="plasma", 
    p0=[0.1, 1, 0, max(aire["theta2"])], # Valores iniciales para gamma, omega, alpha, y A
    save_path= rf'Graficos\ajuste_grafico2.png'
)
```

```python
param_3=ajuste(
    x_data=aire["t3"], 
    y_data=aire["theta3"], 
    modelo_func=third_aire, 
    x_label="Tiempo (s)", 
    y_label="√Ångulo (Œ∏)", 
    etiqueta=r"$\theta$ vs tiempo", 
    modelo_latex=r"$\theta(t) =  0.1e^{-\gamma t} \cos(\omega t + \alpha) + A$",  # Ejemplo de funci√≥n en LaTeX
    color_mapa="Blues", 
    p0=[0.1, 1, 0, max(aire["theta3"])],  # Valores iniciales para gamma, omega, alpha, y A
    save_path= rf'Graficos\ajuste_grafico3.png'
)
```

```python

t = np.linspace(0, 20, 100)  # Dominio para t
theta = np.linspace(0, 2 * np.pi, 100)  # Dominio para theta

# Creando una malla
T, Theta = np.meshgrid(t, theta)
z1 = -13.6 * np.exp(-param_1[0] * T)  # z1
z2 = -9.64 * np.exp(-param_2[0] * T)   # z2
z3 = -4.89 * np.exp(-param_3[0] * T)   # z3

# Coordenadas del s√≥lido de revoluci√≥n
X = T
Y1 = z1 * np.cos(Theta)
Y2 = z2 * np.cos(Theta)
Y3 = z3 * np.cos(Theta)

z1_solid = z1 * np.sin(Theta)
z2_solid = z2 * np.sin(Theta)
z3_solid = z3 * np.sin(Theta)

# Graficar el s√≥lido de revoluci√≥n
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y1, z1_solid, alpha=0.2, cmap='viridis')
ax.plot_surface(X, Y2, z2_solid, alpha=0.8, cmap='pink')
ax.plot_surface(X, Y3, z3_solid, alpha=1, cmap='winter')


# Etiquetas
ax.set_xlabel('Tiempo (s)')
ax.set_ylabel(rf'theta')
ax.set_zlabel(rf'omega')  # Corregido de 'z1' a 'Z'
ax.set_title('Superficies de estados permitidos para cada pendulo')

plt.show()
```
